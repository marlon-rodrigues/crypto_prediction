```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)

kw_list = ['bitcoin']
block_trends = pytrends.get_historical_interest(kw_list, year_start=2013, month_start=10, day_start=1, hour_start=0, year_end=2019, month_end=5, day_end=16, hour_end=23, cat=0, geo='', gprop='', sleep=60)
```

```python
block_trends
```

```python
block_trends.info()
```

```python
import pandas as pd

avg_block_trends = block_trends.groupby(pd.Grouper(freq="D"))[['bitcoin']].mean()
avg_block_trends
```

```python
# Get total number of rows missing a value
avg_block_trends.isna().sum()
```

```python
avg_block_trends.to_csv('block_trends_complete.csv', encoding='utf-8')
```
