```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)

kw_list = ['bitcoin']
# pytrends.build_payload(kw_list, cat=0, timeframe='2013-10-06 2019-05-12', geo='', gprop='')
# cData = pytrends.interest_over_time()
```

```python
blockData = pytrends.get_historical_interest(kw_list, year_start=2019, month_start=4, day_start=1, hour_start=0, year_end=2019, month_end=5, day_end=15, hour_end=23, cat=0, geo='', gprop='', sleep=0)
```

```python
idx_blockData = blockData.reset_index() # reset index so date can be grouped as a column
idx_blockData
```

```python
idx_blockData.info()
```

```python
# Get total number of rows missing a value
idx_blockData.isna().sum()

# Convert NaN to zero (0)
#idx_blockData.fillna(0, inplace=True)
#idx_blockData
```

```python
# Group by day - average values
avg_blockData = idx_blockData.groupby([idx_blockData['date'].dt.year, idx_blockData['date'].dt.month, idx_blockData['date'].dt.day])[['bitcoin']].mean()
avg_blockData
```

```python
import pandas as pd

# BETTER APPROACH
avg2_blockData = blockData.groupby(pd.Grouper(freq="D"))[['bitcoin']].mean()
avg2_blockData
```
