```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
# Read CSV file with Bitcoin data from Kraken - https://www.cryptodatadownload.com/data/northamerican/
bitcoin_data = pd.read_csv('Datasets/Kraken_BTCUSD_d.csv')
bitcoin_data
```

```python
bitcoin_data.info()
```

```python
# Convert date column type from object to date
bitcoin_data['Date'] = pd.to_datetime(bitcoin_data['Date'])
bitcoin_data
```

```python
bitcoin_data.info()
```

```python
len(bitcoin_data)
```

```python
# Check if there are any missing values
bitcoin_data.isna().sum()
```

```python
# Plotting all data
bitcoin_data.set_index('Date').plot(figsize=(15,10), grid=True)
```

```python
# Plotting closing price
bitcoin_data.plot(y='Close', x='Date', figsize=(15,10), grid=True)
```

```python
# Plotting bitcoin volume
bitcoin_data.plot(y='Volume BTC', x='Date', figsize=(15,10), grid=True)
```

```python
# Plotting high and low prices
bitcoin_data.plot(y=['High', 'Low'], x='Date', figsize=(15,10), grid=True)
```

```python
# Plotting open and close prices
bitcoin_data.plot(y=['Open', 'Close'], x='Date', figsize=(20,7), grid=True)
```

```python
# Plotting open, close, high and low prices
bitcoin_data.plot(y=['Open', 'Close', 'High', 'Low'], x='Date', figsize=(20,10), grid=True)
```
